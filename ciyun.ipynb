```python
#绘制词云
from wordcloud import WordCloud
from PIL import Image 
import numpy as np
import re
import matplotlib.pyplot as plt
import jieba
file=open('D:\Jupyter\daode.txt','r',encoding='utf-8')
text=file.read()
file.close()
#print(text)
word_list = jieba.cut(text)
result = " ".join(word_list)
text=result
words=text.split(
#words1=[word.lower() for word in words]  #大写转小写
words2=[re.sub("[，。！；？、\“\”]",'',word) for word in words]#去掉标点符号
words_index=set(words)  #去重复
dic={index:words2.count(index) for index in words_index} 
graph=np.array(Image.open('D:\Jupyter\yun.jfif'))
wc=WordCloud(background_color='white',mask=graph, font_path ='D:\Jupyter\HYQiHei-25JF.ttf',max_words=200,max_font_size=150)
wc.generate_from_frequencies(dic)
wc.to_file("D:\Jupyter\ciyun.jpg")
plt.imshow(wc)
plt.axis("off")
plt.show()
```
